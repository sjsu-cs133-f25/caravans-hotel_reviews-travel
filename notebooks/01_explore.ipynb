#B. Transformations

```python
df.isnull().sum()
```

```python
# At least 1 vectorized boolean mask (e.g., df[col] >= threshold) saved as a new column (e.g., Pass = True/False).

# boolean mask column 'Positive_Tone
df['Positive_Tone'] = df['Review_Total_Positive_Word_Counts'] > df['Review_Total_Negative_Word_Counts']
display(df.head())
```

```python
# At least 1 map or Series.apply used appropriately (single-column transform). - #stay type, hotel address, hotel country
# transformed Hotel_Address column to exclude country usng apply()
df['Hotel_Address'] = df['Hotel_Address'].str.replace('United Kingdom', 'UK')
hotel_country = df['Hotel_Address'].str.split(' ').str[-1]
df['Hotel_Address'] = df.apply(lambda row: ' '.join(row['Hotel_Address'].split(' ')[:-1]), axis=1)
# add country column
df['Hotel_Country'] = hotel_country
# print(df['Hotel_Country'].value_counts())

df.head()
```

```python
# transform days_since_review
df['days_since_review'] = df.apply(lambda row: int(row['days_since_review'].split(' ')[0]), axis = 1)

# print(df['days_since_review'].value_counts())
df.head()
```

```python
# At least 1 categorical bucketing (e.g., bands/labels from a numeric column: low/med/high). -# Reviewer experience, guest type
def avg_score_category(avg_score):
  if avg_score <= 6.00:
    return 'Low'
  elif avg_score > 6.00 and avg_score <= 8.00:
    return 'Medium'
  else:
    return 'High'

df['Average_Score_Category'] = df['Average_Score'].apply(avg_score_category)
df
```

```python
# At least 1 missing-data handling step (drop, fill, or mark) - explain choice briefly in code comments.
# The Hotel_Address and Hotel_Country details are sufficient for the project
# We do not require latitude and longitude details
df.drop(labels=['lat', 'lng'], inplace=True, axis=1)
df.info()
```

```python
# Add a new helper column
df['Reviewer_Difference'] = df.Reviewer_Score - df.Average_Score
df.Reviewer_Difference.head()
```

# C. Grouping & Sorting

```python
import numpy as np

# At least 1 value_counts() with interpretation (not just output).
df.Reviewer_Score.map(np.floor).value_counts().sort_index(ascending=False)
```

### Interpretation (for above):
428887 out of 515738 reviews in this data set, about 83%, give the reviewed hotel a score of 7.0 or higher. This could indicate a bias toward higher scores from reviewers (or perhaps specifically booking.com reviewers), or that the hotels reviewed were simply high quality.

```python
# At least 2 groupby(...).agg(...) queries (e.g., count, nunique, mean, min/max), with sorted results (sort_values).
```

```python
from ast import literal_eval

trip_tags_list = []

# Col 13 is the Tags column, Col 12 is the Reviewer_Score column
for row in df.itertuples():
  for tag in literal_eval(row[13 + 1]):
    trip_tags_list.append([tag, row[12 + 1]])

trip_tags = pd.DataFrame(trip_tags_list, columns=['Tag','Score'])
trip_tags
```

```python
# Mean scores given by tag, sorted highest first
trip_tags.groupby('Tag').Score.agg('mean').sort_values(ascending=False)
# Note that this also contains tags that only occur once, which isn't very useful
```

```python
# Filter to 100 most common tags
tc = trip_tags.Tag.value_counts()
top_trip_tags = trip_tags[trip_tags['Tag'].isin(tc.index[tc.values >= tc.iloc[100 - 1]])]
top_trip_tags.Tag.nunique()
```

```python
# Mean scores given by tag, sorted highest first
top_trip_tags.groupby('Tag').Score.agg('mean').sort_values(ascending=False)
```

```python
# Do some nationalities (culture?) affect ratings?
nc = df.Reviewer_Nationality.value_counts()
reviewer_nationalities = df[df['Reviewer_Nationality'].isin(nc.index[nc.values >= 1000])]
reviewer_nationalities.Reviewer_Nationality.nunique()
```

```python
# Top 5 most positive nations and Top 5 most negative nations (with at least 1000 reviewers)
pd.concat([reviewer_nationalities.groupby('Reviewer_Nationality').Reviewer_Difference.agg('mean').sort_values(ascending=False).head(),
           reviewer_nationalities.groupby('Reviewer_Nationality').Reviewer_Difference.agg('mean').sort_values(ascending=False).tail()])
```

```python
# Either a multi-column groupby or a groupby that you reset_index() and then sort for readability.
hotel_countries = df.Hotel_Country.unique().tolist()
hotel_countries[hotel_countries.index('UK')] = 'United Kingdom'
nationalities = df[df['Reviewer_Nationality'].str.strip().isin(hotel_countries)].groupby(['Hotel_Country', 'Reviewer_Nationality']).Reviewer_Difference.agg('mean')
nationalities
# Could reveal some biases in hotel reviews (own country bias?)
```
